```python
np.random.seed(5)
items_data = np.random.randint(20,size = (5,3))
items_data
```

```python
items_data = pd.DataFrame(items_data,index = ["Mon","Tues","Wed","Thrus","Fri"],columns = ["Almond Butter","Peanut Butter","Cashew Butter"])
items_data
```

```python
prices = np.array([10,8,12])
prices.shape
transposed_price = prices.T
transposed_price.shape #so tranpose do not work 1d arrays

new_prices = prices.reshape(1,3)
new_prices.shape
```

```python
items_price = pd.DataFrame(new_prices,index = ["Price"],columns = ["Almond Butter","Peanut Butter","Cashew Butter"])
items_price
```

```python
total = np.dot(items_data,items_price.T)
total
```

```python
items_data["Total ($)"] = total
print(items_data)
print()
print(items_price)
```

```python
Data = items_data.copy()
price_row = items_price.copy()
price_row.index = ["Price"]
combination = pd.concat([Data,price_row])
combination.to_csv("Butter-sales.csv")
```
