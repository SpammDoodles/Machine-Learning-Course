```python
x = np.linspace(1,10,100)

fig, ax = plt.subplots(figsize = (8,6))

ax.plot(x,x**2)
ax.set(title = "simple line", xlabel = "x-axis",ylabel = "y-axis")
plt.show();
```

```python
fig,ax = plt.subplots(2,1,figsize = (8,10))
ax[0].scatter(x,np.exp(x))
ax[0].set(title = "Scatter Plot using exp",xlabel = "x-axis", ylabel = "y-axis")

ax[1].scatter(x,np.sin(x))
ax[1].set(title = "Scatter Plot using sin",xlabel = "x-axis", ylabel = "y-axis")
plt.tight_layout()
plt.show()
```

```python
nut_butter_prices = {"Almond Butter": 10, "Penut Butter": 8,"Cashew Butter": 12}

fig,ax = plt.subplots(figsize = (8,6))
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title = "My Nut Butter", ylabel = "Prices ($)")
plt.show()
```

```python
fig,ax = plt.subplots(figsize = (8,6))

ax.barh(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title = "My Nut Butter",xlabel = "Prices ($)")
plt.show()
```

```python
random_x = np.random.randn(10)
random_y = np.random.randn(10)
fig,ax = plt.subplots(2,2, figsize = (10,5))
ax[0,0].plot(x,x/2)
ax[0,0].set(title = "plot graph",xlabel = "x-axis", ylabel = "y-axis")
ax[0,1].scatter(random_x,random_y)
ax[0,1].set(title = "scatter chart", xlabel = "x-axis", ylabel = "y-axis")
ax[1,0].barh(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,0].set(title = "barh chart",xlabel = "Prices ($)")
ax[1,1].hist(x)
ax[1,1].set(title = "histogram graph", xlabel = "x-axis",ylabel = "y-axis")
plt.tight_layout()
plt.suptitle("Basic graphs")
plt.show()
```

```python
df = pd.read_csv("datasets/car-sales.csv")

df
```

```python
df["Price"] = df.Price.str.replace(r"[/%#$^.%,]","",regex = True)
df
```

```python
df["Price"] = df["Price"].str[:-2].astype(int)
type(df["Price"][0])
```

```python
df["Total Sales"] = df["Price"].cumsum()
df
```

```python
df["Sales Date"] = pd.date_range("4/3/2009",periods = len(df))
df
```

```python
df.plot(x = "Total Sales",y = "Sales Date");
plt.show()
```

```python
heart_disease = pd.read_csv("datasets/heart-disease.csv")
heart_disease.head(10)
```

```python
heart_disease.plot.hist(x = "age", bins = 50)
plt.show();
```

```python
heart_disease.plot.hist(subplots = True);
plt.show()
```

```python
heart_disease.plot.hist(figsize = (10,30),subplots = True)
plt.show()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(x = "age",y = "chol",c = "target",kind = "scatter",figsize = (10,8))
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,8))
scatter = ax.scatter(x = over_50["age"], y = over_50["chol"],c = over_50["target"])
ax.set(xlabel = "Age",ylabel = "Cholestrol",title = "Heart Disease and Cholestrol Level")
ax.legend(*scatter.legend_elements(),title = "Target")
ax.axhline(over_50["chol"].mean(),linestyle = "--")
cbar = plt.colorbar(scatter,ax = ax)
cbar.set_label("Target");
plt.show()
plt.savefig("images/Heart_disease_and_ch.png");
```
